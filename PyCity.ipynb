```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Store filepath in a variable
schools_data = "raw_data/schools_complete.csv"
students_data = "raw_data/students_complete.csv"

# Use Pandas to read data
schools_df = pd.read_csv(schools_data, encoding = "ISO-8859-1")
students_df = pd.read_csv(students_data, encoding = "ISO-8859-1")
```

```python
# Preview of the DataFrame schools_df
schools_df.head()
```

```python
# Preview of the DataFrame students_df
students_df.head()
```

```python
students_df['mathpass'] = np.where(students_df['math_score']>= 70,1,0)
students_df['readingpass'] = np.where(students_df['reading_score']>= 70,1,0)
students_df.head()
```

```python
# Calculate Total Schools
total_schools =  len(schools_df)
```

```python
# Calculate Total Students
total_students =  len(students_df)
```

```python
# Calculate Total Budget
total_budget = schools_df["budget"].sum()
```

```python
# Calculate Average reading score
avg_reading_score = students_df["reading_score"].mean()
```

```python
# Calculate Average math score
avg_math_score = students_df["math_score"].mean()
```

```python
# % Pass Reading 
perc_passreading = (students_df['readingpass'].sum()/total_students) * 100
```

```python
# % Pass Math
perc_passmath = (students_df['mathpass'].sum()/total_students) * 100
```

```python
# Calculate Overall Pass rate
overall_passrate = (perc_passreading+perc_passmath)/2
```

```python
#Create dataframe for District summary and print in table form

print('\033[1m' + 'District Summary')

District_summary = pd.DataFrame([[(total_schools),'{:,d}'.format(total_students),'${:,.2f}'.format(total_budget),(avg_math_score), (avg_reading_score),(perc_passmath),(perc_passreading),(overall_passrate)]])
District_summary.columns = ['Total Schools','Total Students','Total Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','Overall Passing Rate']
District_summary
```

```python
#create bins
bins=[0,585, 615, 645, 675]
# Create the names for the four bins
spend_labels = ['<$585', '$585-615', '$616-645', '$646-675']
# cut bins according to values
budget_df["Spending Ranges (Per Student)"] = pd.cut(budget_df["Per Student Budget"],bins,labels=spend_labels )
```

```python
budget_groups = budget_df.groupby("Spending Ranges (Per Student)")
print('\033[1m' + "Scores by School Spending")
budget_groups.max()
```
